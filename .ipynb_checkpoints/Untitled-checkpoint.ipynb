```python
### graphical visualisation of the data (0 = dead, 1 = survived)
```

```python
f, ax = plt.subplots(3,4,figsize=(20,14))

sns.countplot('Pclass', data=train, ax=ax[0,0])
sns.countplot('Pclass', hue='Survived', data=train, ax=ax[1,0])
sns.boxplot(x='Pclass', y='Age', data=train, ax=ax[2,0])

sns.countplot('Sex', data=train, ax=ax[0,1])
sns.countplot('Sex', hue='Survived', data=train, ax=ax[1,1])
sns.boxplot(x='Sex', y='Age', data=train, ax=ax[2,1])

sns.countplot('Embarked', data=train, ax=ax[0,2])
sns.countplot('Embarked', hue='Survived', data=train, ax=ax[1,2])
sns.swarmplot(x='Embarked', y='Age', hue='Survived', data=train, ax=ax[2,2])

sns.swarmplot(x='Pclass', y='Age', hue='Survived', data=train, ax=ax[0,3])
sns.countplot('Parch', hue='Survived', data=train, ax=ax[1,3])
sns.countplot('SibSp', data=train, hue='Survived', ax=ax[2,3])
```
